## 2/ Nettoyage des données (Data Cleaning)

On remarque une variation dans la dénomination de la colonne "nom de pays" d'un dataframe à un autre. Afin d'assurer la cohérence et la facilité de manipulation des données, nous avons pris la décision de standardiser cette dénomination en renommant toutes les colonnes correspondantes à "country".
```python
#comparer les pays de df2 et df3
df2 = df2.sort_values('country')
print(df2['country'].unique())
print(df3['country'].unique())
```

On convertit les valeurs en "%" en valeurs numériques réelles.

Normalisation des données avec StandardScaler()

# 5/PCA

On remarque que notre dataframe contient un grand nombre de colonnes, soit un total de 95 colonnes. La gestion de tant de variables peut être complexe et peut entraîner des problèmes de dimensionnalité. Afin de simplifier et de réduire la complexité de nos données, nous avons pris la décision d'utiliser l'analyse en composantes principales (PCA).

```python
from sklearn.decomposition import PCA

# Après des tests on a choisit 15 composants
pca = PCA(n_components=15)
principal_components = pca.fit_transform(data_scaled)

principal_df = pd.DataFrame(data = principal_components, columns = ['PC1', 'PC2','PC3','PC4','PC5','PC6', 'PC7','PC8','PC9','PC10','PC11', 'PC12','PC13','PC14','PC15'])

principal_df
```

```python

# Output de l'ACP, variance expliquée par axe, variance expliquée en %,
# variance cumulée en % pour les 'p' axes de vecteurs directeurs les vecteurs propres 
eig = pd.DataFrame(
    {   "Dimension" : ["Dim" + str(x) for x in range(1,16)], 
        "Varianceexpliquée" : pca.explained_variance_,
        "%variance_expliquée" : np.round(pca.explained_variance_ratio_ * 100),
        "cumvarexpliquée" : np.round(np.cumsum(pca.explained_variance_ratio_) * 100)
    }
)

eig
```

Ainsi, on a pu obtenir un dataframe avec 15 colonnes au lieu de 95. Ces 15 colonnes expliquent 90% de la variance de nos données !

# 6/ Clustering avec Kmeans

Trouvons tout d'abord le nombre optimal de n_clusters

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
dict={}
for i in range(2,11):    
    # intialise kmeans
    kmeans = KMeans(n_clusters=i,random_state=42)
    kmeans.fit(principal_df)
    cluster_labels = kmeans.labels_
     
    dict[i]=silhouette_score(principal_df, cluster_labels)
dict
```

On trouve que le nombre optimal est 2 ou 3 (ils donnent les meilleurs silhouette_score). On choisit de fixer n_clusters à 3. 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(principal_df)

# Le label du cluster pour chaque point
cluster_labels = kmeans.predict(principal_df)

# Ajoutons le label au dataframe
principal_df['cluster'] = cluster_labels

# Le nombre de pays que contient chaque cluster
print(principal_df['cluster'].value_counts())
```

```python
#On ajoute les pays au dataframes

final =pd.concat([X,principal_df],axis=1)
final
```

```python
# On crée une variable qui contient les pays de chaque cluster
Cluster1Countries=final[final['cluster']==0]
Cluster2Countries=final[final['cluster']==1]
Cluster3Countries=final[final['cluster']==2]
```

# 8/ Interprétation des résultats

Dans notre projet de segmentation de pays en utilisant l'algorithme K-means, nous avons obtenu trois clusters distincts. Le premier cluster comprend la Russie et les États-Unis, le deuxième cluster comprend la Chine et l'Inde, tandis que le troisième cluster regroupe tous les autres pays.

Cette segmentation peut être interprétée de la manière suivante : le premier cluster, composé de la Russie et des États-Unis, pourrait représenter les superpuissances mondiales, caractérisées par leur influence politique, économique et militaire significative à l'échelle mondiale.

Le deuxième cluster, qui inclut la Chine et l'Inde, peut être interprété comme regroupant les pays émergents avec de vastes populations et des économies en pleine croissance. Ces pays ont également un rôle de plus en plus important sur la scène internationale.

Enfin, le troisième cluster englobe tous les autres pays, ce qui suggère une segmentation plus diversifiée des pays restants en fonction de leurs caractéristiques socio-économiques, géographiques et politiques spécifiques.
